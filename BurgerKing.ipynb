```python
selected_column = 'Calories'
data_column = df[selected_column]

pdf_values = norm.pdf(data_column, loc=data_column.mean(), scale=data_column.std())

cdf_values = norm.cdf(data_column, loc=data_column.mean(), scale=data_column.std())

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.hist(data_column, bins=30, density=True, alpha=0.6, color='g', label='Histogram')
plt.plot(data_column, pdf_values, 'r-', label='PDF')
plt.title(f' PDF of {selected_column}')
plt.xlabel(selected_column)
plt.ylabel('Probability Density')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(data_column, cdf_values, 'b-', label='CDF')
plt.title(f'CDF of {selected_column}')
plt.xlabel(selected_column)
plt.ylabel('Cumulative Probability')
plt.legend()

plt.tight_layout()
plt.show()

lower_bound = 10
upper_bound = 1220

mean_value = df[selected_column].mean()
std_dev = df[selected_column].std()

probability_within_range = norm.cdf(upper_bound, loc=mean_value, scale=std_dev) - \
                           norm.cdf(lower_bound, loc=mean_value, scale=std_dev)

print(f'Olasılık ({lower_bound} ile {upper_bound} arasında): {probability_within_range:.2%}')
```

Histogram, 'Calories' sütunundaki veri dağınıklığını gösteriyor.
PDF (kırmızı çizgi), normal dağılıma yaklaşan bir olasılık yoğunluk fonksiyonunu temsil eder. Bu bizlere verinin genel dağılımını ifade ediyor.
CDF (mavi çizgi), belirli bir değerin altındaki olasılığı gösterir. Grafiğin yukarı doğru ilerlemesi, bir değerin belirli bir eşiğin altında olma olasılığının arttığını gösteriyor.

Veri Dağılımı:
Histogramdan ve PDF'den görüldüğü üzere, 'Calories' sütunu normal bir dağılıma benziyor. Bu, bizlere verinin genellikle ortalamaya yakın değerlerde yoğunlaştığını gösteriyor.
CDF grafiği, belirli bir kalori değerinin altında olma olasılığını gösterir. Örneğin, CDF'nin belirli bir değerde 0.8 olduğunu göz önüne alırsak, o değerin altında kalori değerine sahip olan öğelerin %80'inin altında olduğunu söyleyebiliriz.
